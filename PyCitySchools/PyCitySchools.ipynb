```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Using len to count the unique values of school name to get the total schools.
Total_Schools = len(school_data_complete["school_name"].unique())
print(Total_Schools)

# Using len to count the unique values of Student ID to get the total students.
Total_Students = len(school_data_complete["Student ID"].unique())
print(Total_Students)

Total_Budget = round(school_data["budget"].sum(),2)
print(Total_Budget)

Average_MathScore = round(student_data["math_score"].mean(),6)
print(Average_MathScore)

Average_ReadingScore = round(student_data["reading_score"].mean(),5)
print(Average_ReadingScore)

Passing_Math_df = student_data.loc[student_data["math_score"] >= 70,:]
Passing_Math = (Passing_Math_df["Student ID"].count() / Total_Students)
print(Passing_Math)

Passing_Reading_df = student_data.loc[student_data["reading_score"] >= 70,:]
Passing_Reading = (Passing_Reading_df["Student ID"].count() / Total_Students)
print(Passing_Reading)

Passing_Overall_df = student_data.loc[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70),:]   
Passing_Overall = (Passing_Overall_df["Student ID"].count() / Total_Students)
print(Passing_Overall)
```

```python
District_Summary_df = pd.DataFrame({
    "Total Schools":  [Total_Schools],
    "Total Students":  [Total_Students],
    "Total Budget":  [Total_Budget],
    "Average Math Score":  [Average_MathScore],
    "Average Reading Score":  [Average_ReadingScore],
    "% Passing Math":  [Passing_Math],
    "% Passing Reading":  [Passing_Reading],
    "% Overall Passing":  [Passing_Overall]
})

District_Summary_df["Total Budget"] = District_Summary_df["Total Budget"].map("${:,.2f}".format)

District_Summary_df["% Passing Math"] = (District_Summary_df["% Passing Math"]*100).map("{:,.6f}".format)
District_Summary_df["% Passing Reading"] = (District_Summary_df["% Passing Reading"]*100).map("{:,.6f}".format)
District_Summary_df["% Overall Passing"] = (District_Summary_df["% Overall Passing"]*100).map("{:,.6f}".format)

District_Summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score

* Create a dataframe to hold the above results

```python
Pre_Index_df = pd.DataFrame({
                                "school_name": [],
                                #"School Type": [],
                                "Total Students": [],
                                "Total School Budget": [],
                                "Per Student Budget": [],
                                "Average Math Score": [],
                                "Average Reading Score": [],
                                "% Passing Math": [],
                                "% Passing Reading": [],
                                "% Overall Passing": []
})

Pre_Index_df
```

```python
for school in school_data["school_name"]:
    
    #print(school)
    
    #School_Type = school_data.loc[(school_data["school_name"] == school), ["type"]]
    School_Type_df = school_data.loc[(school_data["school_name"] == school),:]
    School_Type = School_Type_df["type"]
    #print(School_Type)
    
    Students_PerSchool_df = school_data_complete.loc[(student_data["school_name"] == school),:]
    Students_PerSchool = len(Students_PerSchool_df["Student ID"].unique())
    #print(Students_PerSchool)
    
    Budgets_PerSchool_df = school_data_complete.loc[(student_data["school_name"] == school),:]
    Budgets_PerSchool = Budgets_PerSchool_df["budget"].sum()/Students_PerSchool
    #print(Budgets_PerSchool)    
    
    Budgets_PerStudent_df = school_data_complete.loc[(student_data["school_name"] == school),:]
    Budgets_PerStudent = Budgets_PerSchool/Students_PerSchool
    #print(Budgets_PerStudent)
    
    Average_PerSchool_MathScore_df = student_data.loc[(student_data["school_name"] == school),:]
    Average_PerSchool_MathScore = round(Average_PerSchool_MathScore_df["math_score"].mean(),6)
    #print(Average_PerSchool_MathScore)
    
    Average_PerSchool_ReadingScore_df = student_data.loc[(student_data["school_name"] == school),:]
    Average_PerSchool_ReadingScore = round(Average_PerSchool_ReadingScore_df["reading_score"].mean(),6)
    #print(Average_PerSchool_ReadingScore)
    
    Passing_Math_PerSchool_df = student_data.loc[(student_data["school_name"] == school) & (student_data["math_score"] >= 70),:]
    Passing_Math_PerSchool = (Passing_Math_PerSchool_df["Student ID"].count() / Students_PerSchool)
    #print(Passing_Math_PerSchool)

    Passing_Reading_PerSchool_df = student_data.loc[(student_data["school_name"] == school) & (student_data["reading_score"] >= 70),:]
    Passing_Reading_PerSchool = (Passing_Reading_PerSchool_df["Student ID"].count() / Students_PerSchool)
    #print(Passing_Reading_PerSchool)

    Passing_Overall_PerSchool_df = student_data.loc[(student_data["school_name"] == school) & (student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70),:]   
    Passing_Overall_PerSchool = (Passing_Overall_PerSchool_df["Student ID"].count() / Students_PerSchool)
    #print(Passing_Overall_PerSchool)
    
    to_append = [school, #School_Type, 
                Students_PerSchool, Budgets_PerSchool, Budgets_PerStudent,
                Average_PerSchool_MathScore, Average_PerSchool_ReadingScore,
                Passing_Math_PerSchool, Passing_Reading_PerSchool, Passing_Overall_PerSchool]
    
    series = pd.Series(to_append, index = Pre_Index_df.columns)
    
    Pre_Index_df = Pre_Index_df.append(series, ignore_index=True)

School_Summary_df = Pre_Index_df.sort_values("school_name", ascending=True).set_index(["school_name"])
School_Summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.

  * Combine the series into a dataframe

```python
Pre_gradeavgmath_df = pd.DataFrame({
                                "school_name": [],
                                "9th": [],
                                "10th": [],
                                "11th": [],
                                "12th": [],
})

Pre_gradeavgmath_df
```

```python
for school in school_data["school_name"]:

    Average_9th_MathScore_df = student_data.loc[(student_data["school_name"] == school) & (student_data["grade"] == "9th"),:]
    Average_9th_MathScore = round(Average_9th_MathScore_df["math_score"].mean(),6)
    
    Average_10th_MathScore_df = student_data.loc[(student_data["school_name"] == school) & (student_data["grade"] == "10th"),:]
    Average_10th_MathScore = round(Average_10th_MathScore_df["math_score"].mean(),6)
    
    Average_11th_MathScore_df = student_data.loc[(student_data["school_name"] == school) & (student_data["grade"] == "11th"),:]
    Average_11th_MathScore = round(Average_11th_MathScore_df["math_score"].mean(),6)
    
    Average_12th_MathScore_df = student_data.loc[(student_data["school_name"] == school) & (student_data["grade"] == "12th"),:]
    Average_12th_MathScore = round(Average_12th_MathScore_df["math_score"].mean(),6)
    
    grade_math_append = [school, Average_9th_MathScore, Average_10th_MathScore, Average_11th_MathScore, Average_12th_MathScore]
    
    grade_math_series = pd.Series(grade_math_append, index = Pre_gradeavgmath_df.columns)
    
    Pre_gradeavgmath_df = Pre_gradeavgmath_df.append(grade_math_series, ignore_index=True)

gradeavgmath_df = Pre_gradeavgmath_df.sort_values("school_name", ascending=True).set_index(["school_name"])
gradeavgmath_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
SchoolSum_resetindex_df = School_Summary_df.reset_index()
SchoolSum_resetindex_df.head()
```

```python
Min_avgspending = School_Summary_df["Per Student Budget"].min()
print(Min_avgspending)
Max_avgspending = School_Summary_df["Per Student Budget"].max()
print(Max_avgspending)

bins = [0,585,630,645,680]
bin_labels = ["<$585", "$585-$630", "$630-$645", "$645-$660"]
```

```python
Scores_SchoolSpending_df = SchoolSum_resetindex_df[["Per Student Budget", "Average Math Score", "Average Reading Score",
                                            "% Passing Math", "% Passing Reading", "% Overall Passing"]]

Scores_SchoolSpending_df["Spending Ranges (Per Student)"] = pd.cut(Scores_SchoolSpending_df["Per Student Budget"], bins, labels=bin_labels)

#Scores_SchoolSpending_df.drop(columns=["Per Student Budget"])

Groupby_SchoolSpending_df = Scores_SchoolSpending_df.groupby(["Spending Ranges (Per Student)"]).mean()

#Groupby_SchoolSpending_df.drop(columns="Per Student Budget")

Groupby_SchoolSpending_df["Average Math Score"] = round(Groupby_SchoolSpending_df["Average Math Score"],2)
Groupby_SchoolSpending_df["Average Reading Score"] = round(Groupby_SchoolSpending_df["Average Reading Score"],2)
Groupby_SchoolSpending_df["% Passing Math"] = (Groupby_SchoolSpending_df["% Passing Math"]*100).map("{:,.2f}".format)
Groupby_SchoolSpending_df["% Passing Reading"] = (Groupby_SchoolSpending_df["% Passing Reading"]*100).map("{:,.2f}".format)
Groupby_SchoolSpending_df["% Overall Passing"] = (Groupby_SchoolSpending_df["% Overall Passing"]*100).map("{:,.2f}".format)

Groupby_SchoolSpending_df
```
