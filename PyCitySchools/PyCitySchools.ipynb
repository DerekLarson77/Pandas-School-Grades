```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
Total_Schools = len(school_data_complete["school_name"].unique())
print(Total_Schools)

Total_Students = len(school_data_complete["Student ID"].unique())
print(Total_Students)

Total_Budget = round(school_data["budget"].sum(),2)
print(Total_Budget)

Average_MathScore = round(student_data["math_score"].mean(),6)
print(Average_MathScore)

Average_ReadingScore = round(student_data["reading_score"].mean(),5)
print(Average_ReadingScore)

Passing_Math_df = student_data.loc[student_data["math_score"] >= 70,:]
Passing_Math = (Passing_Math_df["Student ID"].count() / Total_Students)
print(Passing_Math)

Passing_Reading_df = student_data.loc[student_data["reading_score"] >= 70,:]
Passing_Reading = (Passing_Reading_df["Student ID"].count() / Total_Students)
print(Passing_Reading)

Passing_Overall_df = student_data.loc[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70),:]   
Passing_Overall = (Passing_Overall_df["Student ID"].count() / Total_Students)
print(Passing_Overall)
```

```python
District_Summary_df = pd.DataFrame({
    "Total Schools":  [Total_Schools],
    "Total Students":  [Total_Students],
    "Total Budget":  [Total_Budget],
    "Average Math Score":  [Average_MathScore],
    "Average Reading Score":  [Average_ReadingScore],
    "% Passing Math":  [Passing_Math],
    "% Passing Reading":  [Passing_Reading],
    "% Overall Passing":  [Passing_Overall]
})

District_Summary_df["Total Budget"] = District_Summary_df["Total Budget"].map("${:,.2f}".format)

District_Summary_df["% Passing Math"] = (District_Summary_df["% Passing Math"]*100).map("{:,.6f}".format)
District_Summary_df["% Passing Reading"] = (District_Summary_df["% Passing Reading"]*100).map("{:,.6f}".format)
District_Summary_df["% Overall Passing"] = (District_Summary_df["% Overall Passing"]*100).map("{:,.6f}".format)

District_Summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type

  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score

* Create a dataframe to hold the above results

```python
Pre_Index_df = pd.DataFrame({
                                "school_name": [],
                                "School Type": [],
                                "Total Students": [],
                                "Total School Budget": [],
                                "Per Student Budget": [],
                                "Average Math Score": [],
                                "Average Reading Score": [],
                                "% Passing Math": [],
                                "% Passing Reading": [],
                                "% Overall Passing": []
})

Pre_Index_df
```

```python
for school in school_data["school_name"]:
    
    #print(school)
    
    School_Type_df = school_data_complete.loc[(school_data_complete["school_name"] == school),:]
    School_Type = School_Type_df["type"].unique()
    #print(School_Type)
    
    Students_PerSchool_df = school_data_complete.loc[(student_data["school_name"] == school),:]
    Students_PerSchool = len(Students_PerSchool_df["Student ID"].unique())
    #print(Students_PerSchool)
    
    Budgets_PerSchool_df = school_data_complete.loc[(student_data["school_name"] == school),:]
    Budgets_PerSchool = Budgets_PerSchool_df["budget"].sum()/Students_PerSchool
    #print(Budgets_PerSchool)    
    
    Budgets_PerStudent_df = school_data_complete.loc[(student_data["school_name"] == school),:]
    Budgets_PerStudent = Budgets_PerSchool/Students_PerSchool
    #print(Budgets_PerStudent)
    
    Average_PerSchool_MathScore_df = student_data.loc[(student_data["school_name"] == school),:]
    Average_PerSchool_MathScore = round(Average_PerSchool_MathScore_df["math_score"].mean(),6)
    #print(Average_PerSchool_MathScore)
    
    Average_PerSchool_ReadingScore_df = student_data.loc[(student_data["school_name"] == school),:]
    Average_PerSchool_ReadingScore = round(Average_PerSchool_ReadingScore_df["reading_score"].mean(),6)
    #print(Average_PerSchool_ReadingScore)
    
    Passing_Math_PerSchool_df = student_data.loc[(student_data["school_name"] == school) & (student_data["math_score"] >= 70),:]
    Passing_Math_PerSchool = (Passing_Math_PerSchool_df["Student ID"].count() / Students_PerSchool)
    #print(Passing_Math_PerSchool)

    Passing_Reading_PerSchool_df = student_data.loc[(student_data["school_name"] == school) & (student_data["reading_score"] >= 70),:]
    Passing_Reading_PerSchool = (Passing_Reading_PerSchool_df["Student ID"].count() / Students_PerSchool)
    #print(Passing_Reading_PerSchool)

    Passing_Overall_PerSchool_df = student_data.loc[(student_data["school_name"] == school) & (student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70),:]   
    Passing_Overall_PerSchool = (Passing_Overall_PerSchool_df["Student ID"].count() / Students_PerSchool)
    #print(Passing_Overall_PerSchool)
    
    to_append = [school, School_Type, Students_PerSchool, Budgets_PerSchool, Budgets_PerStudent,
                Average_PerSchool_MathScore, Average_PerSchool_ReadingScore,
                Passing_Math_PerSchool, Passing_Reading_PerSchool, Passing_Overall_PerSchool]
    
    series = pd.Series(to_append, index = Pre_Index_df.columns)
    
    Pre_Index_df = Pre_Index_df.append(series, ignore_index=True)

School_Summary_df = Pre_Index_df.sort_values("school_name", ascending=True).set_index(["school_name"])
School_Summary_df
```
